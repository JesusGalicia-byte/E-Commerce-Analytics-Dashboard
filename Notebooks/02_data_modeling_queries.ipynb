```python
import pandas as pd
import sqlite3
%load_ext sql
```

```python
# Load cleaned dataset
df = pd.read_csv("../Data/processed/clean_data.csv")

# Connect to SQLite database
conn = sqlite3.connect('../Data/ecommerce.db')

# Load into SQL table
df.to_sql('orders', conn, if_exists='replace', index=False)
```

```python
# preview of data
query = """
SELECT *
FROM orders
"""

df_view = pd.read_sql_query(query, conn)
df_view
```

KPI's Below

```python
# Sales Overview
query1 = """
SELECT 
    category,
    SUM(quantity) AS total_quantity,
    SUM(total_order_value) AS revenue,
    (SUM(total_order_value) / COUNT(DISTINCT order_id)) AS average_order_value
FROM orders
GROUP BY category
ORDER BY revenue DESC;
"""

df_view = pd.read_sql_query(query1, conn)
df_view
```
